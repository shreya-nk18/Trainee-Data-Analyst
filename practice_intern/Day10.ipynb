# List class methods

```python
list1=[12,67,43,25,99,10,12,12,43]
print(list1.count(12))  #count 12 number of times
print(list1.count(43))
```

```python
list1.index(99)  #checks the index value of a number
```

```python
list1.clear()  #clears the whole list
print(list1)
```

```python
list2=[12,99,18,34,56]
list2.reverse()    #reverse the whole list
print(list2)
```

```python
list2.sort()  #sort the list in ascending order
print(list2)
```

```python
list2.remove(12)  #removes the number from the list
print(list2)
```

```python
list2.pop()   #removes the last element from the list
print(list2)
```

# List Comphrension

```python
#LC is short & smart way to create a new list using a single line of code instead of writing full for loop

[a+1 for a in range(5)]
#0+1 1+1 2+1 3+1 4+1
```

```python
[3+4 for i in range(5)]
#0 1 2 3 4
```

```python
[a**2 for a in range(2,10,2)]
#range- 2,3,4,5,6,7,8,9
#2**2-4
#4**2-16
#6**2-36
#8**2-64
```

```python
[a+3 for a in range(3)]
#0+3 1+3 2+3
```

```python
[a**3 for a in range(1,5,3)]
#range-1,2,3,4
#1**3-1
#4**3-64
```
