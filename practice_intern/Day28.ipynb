# MATPLOTLIB

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
x=[1,2,3,4]
y=[5,6,7,8]
c='hotpink'
plt.plot(x,y,c)
plt.show()
```

```python
x=[2,5,7,8,4]
y=[5,10,23,44,2]
c='g'
plt.plot(x,y,color=c,marker='*')
plt.show()
```

```python
x=[2,5,7,8,4]
y=[5,10,23,44,2]
c='g'
plt.plot(x,y,color=c,marker='o')
plt.show()
```

```python
x=[2,5,7,8,4]
y=[5,10,23,44,2]
c='g'
plt.plot(x,y,color=c,marker='v')
plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y)
plt.plot(x,color='r',marker='o')
plt.plot(y,color='g',marker='*')
plt.show()
```

```python
x=[5,2,9,4,7]
y=[10,5,8,4,2]
plt.plot(x,y)
plt.plot(x,color='r',marker='o')
plt.plot(y,color='y',marker='>' ,ms=15)
plt.show()
```

# BAR GRAPH

```python
x=[1,3,5,7]
y=[9,6,10,8]
c=['darkblue','b','skyblue','lightblue']
plt.bar(x,y,color=c)
plt.show()
```

# BAR GRAPH HORIZONTAL

```python
x=np.array(['A','B','C','D'])
y=np.array([8,4,6,10])
plt.barh(x,y,color='hotpink')
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([10,8,6,4])
plt.barh(x,y,color='skyblue')
plt.show()
```

# VERTICAL

```python
x=np.array(['A','B','C','D'])
y=np.array([10,8,6,4])
plt.bar(x,y,color='skyblue')
plt.show()
```

# WIDTH OF BAR

```python
x=np.array(['A','B','C','D'])
y=np.array([8,4,6,10])
plt.bar(x,y,color='hotpink',width=0.1)
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([8,4,6,10])
plt.bar(x,y,color='skyblue',width=0.8)
plt.show()
```

```python
x=np.array(['A','B','C','D'])
y=np.array([8,4,6,10])
plt.barh(x,y,color='lightblue',height=0.5)
plt.show()
```

```python
x=np.array(['A','B','C','D','E'])
y=np.array([5,7,10,7,5])
plt.bar(x,y,color='r',width=0.9)
plt.show()
```

# SCATTER PLOT

```python
x=[1,2,3,4,5,6]
y=[4,2,3,5,1,8]
plt.title("Scatter plot")  # to put title
plt.xlabel("month")  # to put label
plt.ylabel("Numbers")  # to put label
c=['r','g','hotpink','b','orange','magenta']
plt.scatter(x,y,color=c,s=220)
plt.show()
```

# PIE CHART

```python
x=[10,20,30,40]
y=['Hindi','English','Maths','Science']
c=['y','magenta','brown','blue']
plt.title('Pie chart of marks')
plt.pie(x,labels=y,colors=c)
plt.legend()
plt.show()
```
