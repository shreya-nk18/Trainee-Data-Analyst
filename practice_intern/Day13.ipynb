# Comparison Operators

```python
s1={2,4,6,8}
s2={1,7,3,5}
print(s1<=s2)
print(s1==s2)
print(s1!=s2)
print(s1>=s2)
```

```python
s3={'a','b','c'}
s4={'c','a','b'}
print(s3==s4)
print(s3<s4)
print(s3>s4)
print(s3!=s4)
```
