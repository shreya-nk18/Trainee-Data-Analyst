# Remove vs discard

```python
#remove- It will throw an error bcz value is not available in this set
s1={40,60,11,20,22,30}
s1.remove(100)
```

```python
#Discard - it will not delete anything but will not throw an error
s1={40,60,11,20,22,30}
s1.discard(100)
s1
```

```python
s1.discard(60)
print(s1)
```

```python
s2={12,23,34,45,56,67,78,89}
s2.remove(100)
print(s2)
```

```python
s2.discard(12)
print(s2)
```

```python
s3={"apple","grapes","banana"}
s3.discard("apple")
print(s3)
```

```python
s3.remove("grapes")
print(s3)
```

```python
s4={'a','b','c','d','e','f'}
s4.remove('a')
print(s4)
```

```python
s4.discard('f')
print(s4)
```

```python
s5={10,20,30,40,50}
s5.pop()
print(s5)
```

```python
s6={30,50,20,10,60}
s6.clear()
print(s6)
```

```python
s1={8,10,2}
s2={8,1,5}
s1.intersection(s2)
#it will throw common element,but will return set
```

```python
s1.union(s2)
#it will union the set but return unique value
```

```python
s1.issubset(s2)
#elements should be same in both sets.duplicate values are ignored
```
