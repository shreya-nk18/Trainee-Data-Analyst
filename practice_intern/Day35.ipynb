# Confusion Matrix

```python
import numpy as np
actual=np.array(['Dog','Dog','Dog','Not Dog','Dog','Not Dog','Dog','Dog','Not Dog','Not Dog']) #input
```

```python
predicted=np.array(['Dog','Not Dog','Dog','Not Dog','Dog','Dog','Dog','Dog','Not Dog','Not Dog']) #output
```

```python
actual
```

```python
predicted
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(actual,predicted)  #Syntax: confusion_matrix
cm
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.heatmap(cm,annot=True,xticklabels=["Dog","Not Dog"],yticklabels=["Dog","Not Dog"],cmap='Blues')
```

```python
from sklearn.metrics import classification_report
print(classification_report(actual,predicted))
```
