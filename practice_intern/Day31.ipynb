# Pie Chart

```python
import matplotlib.pyplot as plt
```

```python
x=["Apple","Banana","Mango","Orange"]
y=[150,300,200,250]
plt.xlabel("Name of Fruits")
plt.ylabel("Count of Fruits in Storeroom")
plt.title("The Comparition B/W Fruits and their Count")
plt.pie(y,labels=x)
plt.show()
```

```python
x=["Apple","Banana","Mango","Orange"]
y=[150,300,200,250]
plt.xlabel("Name of Fruits")
plt.ylabel("Count of Fruits in Storeroom")
plt.title("The Comparition B/W Fruits and their Count")
plt.pie(y,labels=x,autopct="%2.3f%%")
plt.show()
```

```python
x=["Apple","Banana","Mango","Orange"]
y=[150,300,200,250]
z=["Red","Yellow","Magenta","Orange"]
plt.xlabel("Name of Fruits")
plt.ylabel("Count of Fruits in Storeroom")
plt.title("The Comparition B/W Fruits and their Count")
plt.pie(y,labels=x,autopct="%2.2f%%",explode=[0,0.1,0,0],shadow=True,colors=z)
plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
age_men = 25,11,68,18,27,28,15,43,58,63,43,65,51,36,33,26,23,35,49,58
plt.hist(age_men, bins=5, color="brown", edgecolor="black")
plt.show()
```

```python
import matplotlib.pyplot as plt
age_men = 25,11,68,18,27,28,15,43,58,63,43,65,51,36,33,26,23,35,49,58
plt.xlabel("Age of Men", color="blue")
plt.ylabel("Frequency of age of men within certain interval", color="blue")
plt.title("Age of men versus frequency", color="red", fontsize=15)
plt.hist(age_men, bins=5, color="brown", edgecolor="black")
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
age_men=25,11,68,18,27,28,15,43,58,63,43,65,51,36,33,26,23,35,49,58
age_women=14,6,2,2,4,15,11,18,1,7,2,1,3,4,10,22,43,45,56,47,55,53,60,
plt.xlabel("Age of Men and Women",color="blue")
plt.ylabel("Frequency of age of men and women within certain interval",color="blue",fontsize=9)
plt.title("Age of men and women versus frequency of age of men and women",color="red",fontsize=15)
plt.hist((age_men,age_women),color=["brown","red"],edgecolor="black",rwidth=0.8)
plt.grid()
plt.legend(["Men","Women"],title="Frequency",loc="upper right")
plt.show()
```

```python
import matplotlib.pyplot as plt
age_men=11,34,66,78,99,66,89,55,78,43,23,89,87,67,90
age_women=67,88,44,56,89,24,85,34,77,32,67,99,53,99,67
age_children=11,22,12,12,12,34,67,88,32,12,33,12,33,12,33,12,12
plt.xlabel("The age of Men and women along with children age",color="blue",fontsize=11)
plt.ylabel("Frequency of age of men,women and children",color="blue",fontsize=11)
plt.title("Age of all 3 genders versus frequency",color="red",fontsize=15)
plt.hist([age_men,age_women,age_children],color=["red","brown","pink"],rwidth=1.3,edgecolor="black",bins=5)
plt.legend(["men","women","children"],title="Frequency",loc="upper right")
plt.grid()
plt.show()
```
