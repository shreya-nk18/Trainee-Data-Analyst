# List

```python
L1=[25,55,11]
for i in L1:
    print(i,end=' ')
```

```python
print(L1[0],L1[-1])
```

```python
A=[21,'shreya',18,'abc',45]
for i in A:
    print(i,end=' ')
```

```python
print(A[-2],A[4])
```

```python
C=["ab","bc","cd"]
for i in C:
    print(i,end=' ')
```

```python
print(C[-2])
```

```python
#List(Delete)
a=[1,2,3,4]
for i in a:
    print(a)
    del a[0]
```

```python
b=["sh","re","ya"]
for i in b:
    print(b)
    del b[1]
```

```python
c=[21,22,23,24,25]
del c[2]
print(c)
```

```python
#List(Edit)
ab=[24,6,78,56]
print(ab)
ab[2]=45
ab[0]=1
print(ab)
```

```python
b=["book","pen","pencil"]
print(b)
b[0]=1
b[2]=3
print(b)
```

```python
#append - add to last position
l1=[1,2,3,4]
print(l1)
l1.append(10)
print(l1)
```

```python
#insert - (position,add)
a=[1,2,3,4]
print(a)
a.insert(4,7)
print(a)
```

```python
a=[1,2,3,4]
print(a)
a.insert(4,7)
del a[1]
print(a)
```
