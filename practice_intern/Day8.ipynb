# Packing & unpacking

```python
l1=[21,43,13]
#unpacking
a,b,c=l1
print(a,b,c)
```

```python
#packing
l2=[b-1,a+2,c+12]
print(l2)
```

```python
l3=[2,54,30,12]
b,c,d,e=l3
print(b,c,d)

l4=[b+2,c-3,d-6]
print(l4)
```

```python
a=[1,2,3,4,5,6]
x,y,z,p,q,r=a
print(x,y,z,p,q,r)

u=[x+1,y+1,z+1,p+1,q+1,r-2]
print(u)
```

```python
list=["x","y"]
a,b=list
print(a,b)

list1=[a,b]
print(list1)
```

```python
list1=[a+"b",b+"c"]
print(list1)
```

```python
#Built-in methods
#len()
l3=[2,54,30,12]
print(len(l3))
```

```python
#min()
print(min(l3))
```

```python
#max
print(max(l3))
```

```python
#sum
print(sum(l3))
```

```python
a=[12,2,23,4]
print(len(a))
print(min(a))
print(max(a))
print(sum(a))
```

```python
#sorted
print(sorted(a))
```

```python
#reverse the list
print(sorted(a,reverse=True))
```

```python
my_list=[43,25,44,55,76,89,78]
print(sorted(my_list))
print(sorted(my_list,reverse=True))
```

```python
l4=[12,98,45,10,2,43,87]
print(len(l4))
print(min(l4))
print(max(l4))
print(sum(l4))
print(sorted(l4))
print(sorted(l4,reverse=True))
```
