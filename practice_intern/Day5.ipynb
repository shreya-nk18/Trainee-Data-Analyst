# Match_case

```python
def http_status(code,a,b):
    match code:
        case '+':
            return a+b
        case '-':
            return a-b
        case '*':
            return a*b
        case '/':
            return a/b
        case '%':
            return a%b
print(http_status("+",2,2))
print(http_status("-",2,2))
```

```python
def http_status(code,a,b):
    match code:
        case 1:
            return a+b
        case 2:
            return a-b
        case 3:
            return a*b
        case 4:
            return a/b
        case 5:
            return a%b
print(http_status(1,2,2))
print(http_status(4,2,2))
```
