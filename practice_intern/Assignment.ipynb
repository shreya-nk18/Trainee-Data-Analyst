# 3. Data visualization

```python
#Histogram
import matplotlib.pyplot as plt
df.hist()
plt.tight_layout()
plt.show()
```

```python
#Boxplot
import seaborn as sns
plt.figure(figsize=(12,6))
sns.boxplot(data=df)
plt.tight_layout()
plt.show()
```

# Bar chart

```python
plt.figure(figsize=(8,5))

df['NSP'].value_counts().plot(kind='bar')

plt.xlabel("Fetal State")
plt.ylabel("Count")
plt.title("Frequency of Fetal States")
plt.show()
```

```python
# pie chart
df['NSP'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Fetal State Distribution")
plt.ylabel("")
plt.show()
```

# Scatter plot

```python
plt.scatter(df['LB'], df['AC'])
plt.xlabel("Baseline Heart Rate (LB)")
plt.ylabel("Accelerations (AC)")
plt.title("LB vs AC")
plt.show()
```

```python
# correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

# Pair plot

```python
sns.pairplot(df)
plt.show()
```
