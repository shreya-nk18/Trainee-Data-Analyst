# write the python script to print 1 to 10

```python
for i in range(1,11):
    print(i,end=' ')
```

# write python script to print range (1,23,3)

```python
for i in range(1,23,3):
    print(i,end=' ')
```

# print first 10 multiples of 5

```python
for i in range(1,11):
    print(5*i,end=' ')
```

# first 10 multiples of n

```python
n = int(input())
for i in range(1, 11): 
    print(n*i,end=' ')
```

# first m multiples of n

```python
n, m = map(int, input().split())
for i in range(1, m+1): 
    print(n*i,end=' ')
```

# first 10 multiples of n in reverse order

```python
n = int(input())
for i in range(10,0,-1): 
    print(n*i,end=' ')
```

# print table of users choice

```python
n = int(input("Enter number: "))
for i in range(1, 11):
    print(n, "x", i, "=", n*i)
```
