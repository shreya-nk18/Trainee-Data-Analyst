# Slicing Opeartors

```python
l1=[45,67,12,32,10,98,70,31]
print(l1[1:6:2])
#[beg:end:step]
#beg-inclusive, end-exclusive
```

```python
l2=["ab","bc","cd","de"]
print(l2[0:7:2])
```

```python
l3=[1,2,3,4,5,6,7,8]
print(l3[0:8:3])
```

```python
l4=['a','c','e','w']
print(l4[0:5:2])
```
