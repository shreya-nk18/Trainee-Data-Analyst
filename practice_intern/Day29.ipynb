# Merging & Joining

```python
import pandas as pd
employee=pd.DataFrame({"EmpID":[1,2,3],
                      "Name":["Alice","Bob","Charlie"],
                      "DeptID":[10,20,30]})
employee
```

```python
department=pd.DataFrame({"DeptID":[10,20,30],
                      "DeptName":["HR","Engineering","Marketing"],
                      })
department
```

```python
pd.merge(employee,department,on="DeptID")
```

```python
pd.merge(employee,department,how="inner",on="DeptID")
```

```python
pd.merge(employee,department,how="left",on="DeptID")
```

```python
pd.merge(employee,department,how="right",on="DeptID")
```

```python
pd.merge(employee,department,how="outer",on="DeptID")
```
