# Box Plot

```python
import pandas as pd
df=pd.read_csv("tips.csv")
df
```

```python
df.shape
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.boxplot(x="day",y="total_bill",data=df)
plt.show()
```

```python
sns.boxplot(x="day",y="total_bill",data=df)
sns.set_style("dark")
plt.show()
# (white,dark,darkgrid,whitegrid)
```

```python
# INFER:people are visiting on sunday more than other days and people are spending more ammount on sunday than other days
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x="day",y="size",data=df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x="sex",y="size",data=df)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x="day",y="total_bill",data=df,color="red")
sns.set_style("dark")
#adding single color
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.boxplot(x="day",y="total_bill",data=df,palette=["red","green","blue","yellow"])
sns.set_style("dark")
#adding multiple color
# we can add plt.xlabel and ylable and title
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("dark")

# Boxplot with future-safe palette usage
sns.boxplot(x="day", y="total_bill", hue="day", data=df,
            palette=["red", "green", "blue", "yellow"])
plt.show()
```

# Line Plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,15,25,30]
plt.plot(x, y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,15,25,30]
plt.plot(x, y, color="blue", marker="o", linestyle="--")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,15,25,30]
plt.xlabel("Days")
plt.ylabel("Sales")
plt.title("Sales Over Time")
plt.plot(x, y, color="blue", marker="o", linestyle="--")
plt.show()
```
