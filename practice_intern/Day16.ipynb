# Normal func

```python
def square(n):
    return(n**2)
arr=square(5)
arr
```

# Lambda Func

```python
square=lambda x:x**2
square(5)
```

# Map Func

```python
number=[1,2,3,4]
square=list(map(lambda x:x**2,number))
                    #arg       #arg
#atleast 2 arguments should be there
square
```

# Filter func

```python
number=[1,2,3,4]
evens=(filter(lambda x:x%2==0,number))
print(list(evens))
```

# swap of map & filter

```python
number=[1,2,3,4]
square=list(filter(lambda x:x**2,number))
square
```

```python
number=[1,2,3,4]
evens=(map(lambda x:x%2==0,number))
print(list(evens))
```

# Reduce func

```python
from functools import reduce
number=[1,2,3,4]
product=reduce(lambda x,y:x*y,number)
print(product)
```

```python
add=lambda x,y:x+y
add(3,4)
```

```python
max=lambda x,y:x if x>y else y
max(5,10)
```

# Sorting

```python
#student[0]-name
#student[1]-marks
student=[("alice",85),("bob",97),("charlie",92)]
student.sort(key=lambda student:student[0])
student
```

```python
student=[("alice",85),("bob",97),("charlie",92)]
student.sort(key=lambda student:student[0],reverse=True)
student
```
