# DEFAULT ARGUMENTS

```python
def add(a,b):
    return(a+b)
def main():
    print("sum of 2,3,5 is",add(2,3,5))
main()
```

```python
def add(a,b):
    return(a*b)
def main():
    print("sum of 2,5 is",add(2,5))
main()
```

```python
def add(a,b,c=0):
    return a+b+c
def add1(a,b,c):
    return a+b
def main():
    print("sum of 10,20,30 is",add(10,20,30))
    print("sum of 20 and 30 is",add1(20,30))
main()
```

```python
def add(a,b,c=0):
    return a+b+c
def main():
    print("sum of 2,5 is",add(2,5))
    print("sum of 10,20 and 30 is",add1(10,20,30))
main()
```

```python
def add(a=0,b=0,c=0):
    return a+b+c
def main():
    print("sum of nothing",add())
    print("sum of 10,20 and 30 is",add1(10,20,30))
main()
```
