# Pivot

```python
import pandas as pd
data={"Name":["Alice","Bob","Charlie"],"Math":[89,78,87],"Science":[56,67,65],"English":[87,67,56]}
```

```python
df=pd.DataFrame(data)
df
```

```python
df2=df.melt(id_vars=["Name"],value_vars=["Math","Science","English"],var_name="Subject",value_name="Score")
df2
```

```python
df3=df2.pivot(index="Name",columns="Subject",values="Score")
df3
```

```python
df3=df2.pivot(index="Name",columns="Subject",values="Score").reset_index()
df3
```

```python
df3.columns.name=None
df3
```

```python
data={"Name":["Alice","Alice","Alice","Bob"],"Subject":["Math","Science","Math","Science"],
"Score":[80,90,70,65]}
df4=pd.DataFrame(data)
df4
```

```python
df4.pivot_table(index="Name",columns="Subject",values="Score",aggfunc="mean")
```

# Group By

```python
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
}) 
df
```

```python
df.info()
```

```python
df.groupby("Department")
```

```python
df.groupby("Department")["Salary"].mean()
```

```python
df.groupby("Team")["Salary"].mean()
```

```python
df.groupby("Team")["Salary"].max()
```

```python
df.groupby("Team")["Salary"].min()
```

```python
df.groupby("Team")["Salary"].count()
```

```python
df.groupby("Team")["Salary"].sum()
```

```python
df.groupby(["Team","Gender"])["Salary"].max()
```

```python
df.groupby("Team")["Salary"].agg(["mean","max","min"])
```

```python
df.groupby("Team").agg({"Salary":"mean","Age":"max"})
```

```python
df["Team Average"]=df.groupby("Team")["Salary"].agg("mean")
df
```

```python
df.groupby("Team")["Salary"].agg("mean")
```

```python
df.groupby("Team")["Salary"].transform("mean")
```

```python
df["Team Average"]=df.groupby("Team")["Salary"].transform("mean")
df
```

```python
df.groupby("Team").filter(lambda x:x["Salary"].mean()>80)
```
