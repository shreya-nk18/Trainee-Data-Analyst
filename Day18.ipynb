# Changing properties

```python
class person():  #initailizing class[creating aclass named person]
    name="shreya"   #properties of class
    age=21          #properties of cls
    def info(self):    #(self) should be the default
        print(f"{self.name} is {self.age}")
a=person()   #object
print(a.info())
```

```python
b=person()
b.name="bob"
b.age=25
b.info()
```

# Self

```python
class person():
    def __init__(self):
        print("Hello")
a=person()
```

```python
class person():
    def __init__(self,name,age):
        self.name=name
        self.age=age
    def info(self):
        print(f"{self.name} is a {self.age}")
a=person("shreya",21)
```

```python
a.info()
```

# Execution time for list

```python
import time
size=10_00_000
l1=list(range(size))
l2=list(range(size))
start=time.time()
l3=[x*y for x,y in zip(l1,l2)]
end=time.time()
print(l3[0:10])
print(f"Time taken by a list is {end-start}")
```

```python
import time
import numpy as np
size=10_00_000
arr1=np.array(list(range(size)))
arr2=np.array(list(range(size)))
start=time.time()
arr3=arr1*arr2
end=time.time()
print(l3[0:10])
print(f"Time taken by a list is {end-start}")
```
